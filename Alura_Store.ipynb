### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
total = loja['Preço'].sum()
loja['Faturamento'] = total
loja.head()
dados = {
    'Arquivo': ['Loja1'],
    'Faturamento': [total]
}
df1 = pd.DataFrame(dados)
df1.to_csv('Loja1_total.csv',index=False,sep=';')
df1.head()


```

```python
total = loja2['Preço'].sum()
loja2['Faturamento'] = total
loja2.head()
dados = {
    'Arquivo': ['Loja2'],
    'Faturamento': [total]
}
df2 = pd.DataFrame(dados)
df2.to_csv('Loja2_total.csv',index=False,sep=';')
df2.head()

```

```python
total = loja3['Preço'].sum()
loja3['Faturamento'] = total
loja3.head()
dados = {
    'Arquivo': ['Loja3'],
    'Faturamento': [total]
}
df3 = pd.DataFrame(dados)
df3.to_csv('Loja3_total.csv',index=False,sep=';')
df3.head()
```

```python
total = loja4['Preço'].sum()
loja4['Faturamento'] = total
loja4.head()
dados = {
    'Arquivo': ['Loja4'],
    'Faturamento': [total]
}
df4 = pd.DataFrame(dados)
df4.to_csv('Loja4_total.csv',index=False,sep=';')
df4.head()
```

```python
df_horizontal = pd.concat([df1, df2,df3,df4], axis=0)
df_horizontal.to_csv('total_faturamento.csv',index=False,sep=';')
grafico = df_horizontal.plot(kind='barh', x = 'Arquivo', y = 'Faturamento',color='blue',legend=False,figsize=(10,5))

grafico.set_title('Alura Store')
grafico.set_xlabel('Faturamento')
grafico.set_ylabel('Lojas')

```
